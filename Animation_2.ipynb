```python
__author__ = '@Tssp'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
plt.rcParams['xtick.labelsize']=16
plt.rcParams['ytick.labelsize']=16
plt.rcParams['axes.labelsize']=19
plt.rcParams['axes.titlesize']=19
```

```python
def f(w, t, k, x):
    return np.sin(w*t - k*x)

w = 2*np.pi
L = 1
m = 3
k = np.pi*m/L
x = np.linspace(0, 1)
```

```python
def init():
    ax.set_xlim(0, 1)
    ax.set_ylim(-2.1, 2.1)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    plt.tight_layout()
    plt.legend(['$y_1$', '$y_2$', '$y_3$'], loc='lower left')
    return ln,
```

```python
def update(frame):
    xdata.append(frame)
    ydata = f(w, frame, k, x)
    ydata2 = f(w, frame, -k, x)
    ln.set_data(x, ydata)
    ln2.set_data(x, ydata2)
    ln3.set_data(x, ydata - ydata2)
    return ln,
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r', alpha=.5)
ln2, = plt.plot([], [], 'b', alpha=.5)
ln3, = plt.plot([], [], 'g', lw=2)

ani = animation.FuncAnimation(fig, update, frames=np.linspace(0, 10, 120),
                    init_func=init, blit=True, interval = 100)

ani.save(f'Interferencia_m{m}.gif', writer='imagemagick', dpi=200)
plt.show()
```
